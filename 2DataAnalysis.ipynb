# Finding Most Common Word

```python
import pandas as pd
```

```python
file='heart_dark.txt'
```

```python
def parse_data(file):
    f=open(file,'r')
    data=f.read().lower()
    cleanup='. , ! ? " '.split()
    for mark in cleanup:
        data.replace(mark,'')
    return data.split()
```

```python
data=parse_data(file)
```

```python
data
```

```python
words=pd.DataFrame(data)
```

```python
words.head()
```

```python
 words.info()
```

```python
words.rename(columns={0:'Word'},inplace=True)
```

```python
words.groupby('Word').size().sort_values(ascending=False)
```

```python
import yfinance as yf

# Download NVIDIA stock data from 2018-10-01
nvda = yf.download('NVDA', start='2018-10-01')

# Display the first few rows
print(nvda.head())
```

```python
nvda.info()
```

```python
nvda.head()
```

```python
nvda=round(nvda,2)
nvda['Change'] = (nvda['Close'] - nvda['Close'].shift(1)).fillna(0)
```

```python
nvda.head()
```

```python
import numpy as np
```

```python
nvda['Return']=np.log(nvda['Close']).diff()
nvda.head()
```

```python
nvda['Close'].pct_change()
```

```python
nvda['Volatility']=nvda['Return'].rolling(21).std().shift(1)
nvda.head(10)
```

```python
nvda.dropna(inplace=True)
nvda.head()
```

```python
print(nvda['Volatility'].shape)
print(nvda['Close'].shape)
```

```python
nvda['Exp Change'] = nvda['Volatility'].squeeze() * nvda['Close'].squeeze()
nvda['Move Size'] = nvda['Change'].squeeze() / nvda['Exp Change']
# .squeeze() converts a (n,1) DataFrame into a 1D Series.
```

```python
nvda.head()
```
